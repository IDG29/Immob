```python
# Locate the output directory
output_folder="/home/idg29/BSC/Immob/Proyectos/Try/0042/ProMPNN_jobs"
output_path = os.path.join(output_folder, "output")

if not os.path.exists(output_path):
    raise ValueError(f"The folder 'output' was not found in {output_folder}")

# Iterate over model subfolders in the output directory
metadata = {}
pdb_name_list = []
chains_list = []
score_list = []
global_score_list = []
for model_folder in os.listdir(output_path):
    model_path = os.path.join(output_path, model_folder)
    seq_path = os.path.join(model_path, f"seqs/{model_folder}.fa")

    
    
    with open(seq_path) as f:
        lines = f.readlines()
    
        for line in lines:
            line = line.strip()
            
            if re.match(r"^>\d+", line):
                pdb_name = (line.split(","))[0]
                pdb_name_list.append(pdb_name)
                designed_chains = (line.split(","))[4]
                chains_list.append(designed_chains)
                score = float(((line.split(","))[1]).split("=")[1])
                score_list.append(score)
                global_score = float(((line.split(","))[2]).split("=")[1])
                global_score_list.append(global_score)
            
                metadata[pdb_name] = {
                    "score" : score,
                    "global_score" : global_score
                    
                }


data = {
    "pdb_name" : pdb_name_list,
    "chain" : chains_list,
    "score" : score_list,
    "global_score" : global_score_list   
}

df = pd.DataFrame(data)
df.replace(to_replace="nan", value=np.nan, inplace=True )
df.dropna(inplace=True)
df.sort_values(by="score", ascending=False)
    



    
```
